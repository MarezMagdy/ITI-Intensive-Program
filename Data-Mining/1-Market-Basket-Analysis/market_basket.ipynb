```python
df = pd.read_excel("Egyptian_Grocery_Transactions.xlsx" , header=None)
```

```python
df.columns = ['items']
```

```python
transactions = df['items'].str.split(',').tolist()
```

```python
transactions
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
t1 = TransactionEncoder()
trans = t1.fit_transform(transactions)

df_encoded = pd.DataFrame(data=trans , columns=t1.columns_)
```

```python
df_encoded
```

```python
support = apriori(df_encoded , min_support=0.1, use_colnames=True )
support
```

```python
support.sort_values('support', ascending= False)
```

```python
rules = association_rules(support , metric= 'confidence', min_threshold= 0.3)
rules = rules.sort_values('confidence', ascending= False)
rules
```
